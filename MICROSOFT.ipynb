# DATA CLEANING
## 1. MOVIE_BUDGETS_DF

```python
#CONVERTING COLUMNS TO FLOATS. CHANGE COLUMN NAME FOR 'worldwide_gross' TO 'foreign_gross'
movie_budgets['production_budget'] = movie_budgets['production_budget'].str.replace("$","",regex=True).str.replace(",","").astype(str).astype(int)
movie_budgets['domestic_gross'] = movie_budgets['domestic_gross'].str.replace("$","",regex=True).str.replace(",","").astype(str).astype(int)
movie_budgets['worldwide_gross'] = movie_budgets['worldwide_gross'].str.replace("$","",regex=True).str.replace(",","").astype(str).astype(float)
```

```python
# CHECKING FOR NULL VALUES
movie_budgets.isnull().values.any()
```

```python
# CHECKING FOR DUBLICATED VALUES
movie_budgets.duplicated().any()
```

```python
# DROP UNWANTED COLUMN 'id' AND RENAME COLUMN 'movie' to 'title'
# movie_budgets = movie_budgets.drop(columns = ['id'])
movie_budgets = movie_budgets.rename(columns = {'movie':'title'})
```

## CORRELATION OF OTHER ELEMENTS AND PRODUCTION COST

Production budget is weakly positively correlated with runtime minutes and average rating. ***Higher production cost does not imply more runtime minutes and increased rating.***. Production budget is highly correlated with domestic gross and world wide gross. ***This implies that, the more cost incured during the production, the higher the domestic and worldwide gross hence more profit***. It can also be noted that domestic gross is highly positively correlated with worldwide gross. An increase in domestic gross leads to an increase in worldwide gross. 

```python
corr_values = merged[['production_budget','runtime_minutes','averagerating','domestic_gross','worldwide_gross']]
corr = sns.heatmap(corr_values.corr(),cmap="YlGnBu", annot=True)
plt.title("Correlation")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (20,15) 

plt.subplot(2,2,1)
x1=merged['production_budget']
y1=merged['runtime_minutes']
plt.scatter(x1,y1,c='r')
plt.title('Correlation between runtime minutes and production_budget')
plt.grid()


plt.subplot(2,2,2)
x2=merged['production_budget']
y2=merged['averagerating']
plt.scatter(x2,y2,c='g')
plt.title('Correlation between average rating and production_budget')
plt.grid()


plt.subplot(2,2,3)
x3=merged['domestic_gross']
y3=merged['production_budget']
plt.scatter(x3,y3,c='b')
plt.title('Correlation between domestic gross and production_budget')
plt.grid()


plt.subplot(2,2,4)
x4=merged['production_budget']
y4=merged['worldwide_gross']
plt.scatter(x4,y4,c='y')
plt.title('Correlation between worldwide gross and production_budget')
plt.grid()
```
